### Explore Bike Share Data

```r
# Importing packages
library(dplyr)
library(ggplot2)
```

```r
# Reading the csv's and inserting a column with the city name for when we merge the data sets later.
chi <- read.csv("chicago.csv") %>% mutate(City = "Chicago")
ny <- read.csv("new_york_city.csv") %>% mutate(City = "New York City")
wash <- read.csv("washington.csv") %>% mutate(City = "Washington")
```

```r
# Merging the data for all three cities into one dataset, ignoring rows where User.Type isn't Customer or Subscriber
bike_data <- bind_rows(chi, ny, wash) %>%
  filter(User.Type %in% c("Customer", "Subscriber"))
```

```r
bike_data
```

```r
# Creating a vector of colorblind-friendly colors to use for city grouping in visualizations
cb_pal <- c("#0072B2", "#E69F00", "#009E73")
```

### Question 1

What are the most popular routes where the start and end stations are *NOT* the same?

```r
# Create a new data frame that includes only non-round-trip trips
non_round_trips <- bike_data %>% filter(Start.Station != End.Station)

# Use group_by and summarize to count the number of trips for each unique route
route_counts <- non_round_trips %>% 
  group_by(City, Start.Station, End.Station) %>% 
  summarize(TripCount = n()) %>% 
  arrange(desc(TripCount))

# Select the top 5 routes for each city
top_routes <- route_counts %>% 
  group_by(City) %>% 
  top_n(5, TripCount)
```

```r
# Create a bar chart showing the top 5 routes for each city
ggplot(top_routes, aes(x = reorder(paste(Start.Station, " - ", End.Station), -TripCount), y = TripCount, fill = City)) +
  geom_bar(stat = "identity", position = "dodge") +
  geom_text(aes(label = TripCount), vjust = -0.5, size = 3.5, color = "black") +
  ggtitle("Top 5 Non-Round-Trip Routes by City") +
  xlab("Route") +
  ylab("Trip Count") +
  scale_fill_manual(values = cb_pal) +
  theme(axis.text.x = element_text(angle = 75, hjust = 1, size = 7),
        axis.text.y = element_text(size = 10),
        axis.title = element_text(size = 12),
        plot.title = element_text(size = 14))
```

**Question 1 Summary:**
* Washington has the highest number of non-round-trip routes, with New York in second, and Chicago with the least.

```r
by(route_counts$TripCount, route_counts$City, function(x) summary(x[order(x, decreasing = TRUE)][1:5]))
```

### Question 2

What percentage of trips in each city were taken by subscribers vs. customers?

```r
# Use group_by and summarize to count the number of trips for each user type in each city, ignoring rows with missing User.Type values
user_counts <- bike_data %>% 
  filter(!is.na(User.Type)) %>%
  group_by(City, User.Type) %>% 
  summarize(TripCount = n()) %>% 
  ungroup()

# Use group_by and summarize to count the total number of trips in each city
total_trips <- bike_data %>% 
  group_by(City) %>% 
  summarize(TotalTrips = n()) %>% 
  ungroup()

# Merge the user counts and total trips data frames
user_percentages <- left_join(user_counts, total_trips, by = "City") %>% 
  mutate(Percentage = round(TripCount / TotalTrips * 100, 2))
```

```r
# Create a bar chart showing the percentage of trips taken by subscribers and customers in each city
ggplot(user_percentages, aes(x = City, y = Percentage, fill = User.Type)) +
  geom_bar(stat = "identity", position = "dodge") +
  ggtitle("Percentage of Trips by User Type and City") +
  xlab("City") +
  ylab("Percentage of Trips") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 10),
        axis.text.y = element_text(size = 10),
        axis.title = element_text(size = 12),
        plot.title = element_text(size = 14))
```

**Question 2 Summary:**

```r
# Display a statistical summary of the percentage of trips taken by subscribers and customers in each city
user_percentages %>%
  group_by(User.Type) %>%
  summarize(
    Mean = mean(Percentage),
    Median = median(Percentage),
    Min = min(Percentage),
    Max = max(Percentage),
    `1st Quartile` = quantile(Percentage, 0.25),
    `3rd Quartile` = quantile(Percentage, 0.75),
    `Standard Deviation` = sd(Percentage)
  ) %>%
  knitr::kable(digits = 2)
```

### Question 3

Does the amount of bike stations for each city correlate to a higher average number of trips?

```r
# Create a data frame with the number of trips and stations in each city
city_summary <- bike_data %>%
  group_by(City) %>%
  summarize(TotalTrips = n(), TotalStations = n_distinct(Start.Station)) %>%
  ungroup() %>%
  mutate(TripsPerStation = TotalTrips / TotalStations)

# Generate a pie chart of the average trips per station for each city
ggplot(city_summary, aes(x = "", y = TripsPerStation, fill = City)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar("y", start = 0) +
  labs(
    title = "Average Trips per Station in Each City",
    x = NULL,
    y = "Average Trips per Station",
    fill = "City"
  ) +
  scale_fill_manual(values = cb_pal) +
  theme_void()
```

**Summary of your question 3 results goes here.**

```r
by(city_summary$TripsPerStation, city_summary$City, summary)
```
